```python
import pandas as pd
import numpy as np
df = pd.read_csv('data.csv')
df.set_index(['Country', 'Year'], inplace=True)
print("Original DataFrame:\n", df, "\n")
print("Mean of Population:", df['Population'].mean())
print("Sum of GDP:", df['GDP'].sum())
print("Max Life Expectancy:", df['Life_Expectancy'].max())
print("np.mean(Population):", np.mean(df['Population']))
print("np.std(GDP):", np.std(df['GDP']))
print("np.sum(GDP):", np.sum(df['GDP']))
print("np.min(Life_Expectancy):", np.min(df['Life_Expectancy']), "\n")
df['GDP'] = df['GDP'].apply(lambda x: x * 1.10)
print("After Increasing GDP by 10%:\n", df, "\n")
df_swapped = df.swaplevel().sort_index()
print("After swaplevel (Year, Country):\n", df_swapped, "\n")
df_unstacked = df.unstack(level='Year')
print("Unstacked DataFrame:\n", df_unstacked, "\n")
print("Population trend by Country and Year:\n")
for country in df.index.get_level_values(0).unique():
    print(f"{country}:\n", df.loc[country, 'Population'], "\n")
```

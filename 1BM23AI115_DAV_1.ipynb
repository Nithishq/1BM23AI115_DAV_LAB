```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_dataset.csv")  # Replace with actual filename

# Ensure Salary_LPA has no NaNs for plotting
df["Salary_LPA"] = df["Salary_LPA"].fillna(0)

# Basic Plot
plt.figure()
plt.plot(df["Age"], df["CGPA"], marker='o')
plt.title("Age vs CGPA")
plt.xlabel("Age")
plt.ylabel("CGPA")
plt.grid(True)
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(df["CGPA"], df["Salary_LPA"], c='green')
plt.title("CGPA vs Salary")
plt.xlabel("CGPA")
plt.ylabel("Salary (LPA)")
plt.show()

# Bar Plot
mean_cgpa_by_dept = df.groupby("Department")["CGPA"].mean()
plt.figure()
mean_cgpa_by_dept.plot(kind='bar', color='skyblue')
plt.title("Average CGPA by Department")
plt.ylabel("Average CGPA")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Horizontal Bar
plt.figure()
mean_cgpa_by_dept.plot(kind='barh', color='orange')
plt.title("Average CGPA by Department")
plt.xlabel("Average CGPA")
plt.tight_layout()
plt.show()

# Histogram
plt.figure()
plt.hist(df["Study_Hours_Per_Week"], bins=10, color='purple')
plt.title("Study Hours Per Week")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Boxplot
plt.figure()
plt.boxplot([df["CGPA"], df["Attendance_Percent"], df["Salary_LPA"]], labels=["CGPA", "Attendance", "Salary"])
plt.title("Boxplots of Various Attributes")
plt.grid(True)
plt.show()

# Violin plot
plt.figure()
plt.violinplot([df["CGPA"], df["Study_Hours_Per_Week"]], showmeans=True)
plt.title("Violin Plot: CGPA & Study Hours")
plt.xticks([1, 2], ["CGPA", "Study Hours"])
plt.show()

# Pie chart
placed_counts = df["Placed"].value_counts()
plt.figure()
plt.pie(placed_counts, labels=placed_counts.index, autopct="%1.1f%%", startangle=140)
plt.title("Placement Distribution")
plt.axis('equal')
plt.show()

# Stackplot
x = np.arange(len(df))
plt.figure()
plt.stackplot(x, df["Study_Hours_Per_Week"], df["Internship_Months"], labels=["Study Hours", "Internships"])
plt.legend(loc='upper left')
plt.title("Stackplot: Study Hours & Internships")
plt.show()

# Stem plot
plt.figure()
markerline, stemlines, baseline = plt.stem(df["Projects_Count"].values[:15], basefmt="r-")
plt.title("Stem Plot: Project Counts (first 15)")
plt.xlabel("Student Index")
plt.ylabel("Projects")
plt.show()

# Errorbar plot
mean = df["CGPA"].mean()
std = df["CGPA"].std()
x = np.arange(1, 6)
y = [mean] * 5
errors = [std] * 5
plt.figure()
plt.errorbar(x, y, yerr=errors, fmt='-o')
plt.title("Errorbar: CGPA Mean  Std Dev")
plt.show()

# Fill Between
x = df["Age"].sort_values()
y1 = df.sort_values("Age")["CGPA"]
plt.figure()
plt.plot(x, y1, color='blue')
plt.fill_between(x, y1, alpha=0.3)
plt.title("Fill Between: Age vs CGPA")
plt.show()

# Image Plot using imshow()
matrix = df.select_dtypes(include='number').corr().values
plt.figure()
plt.imshow(matrix, cmap="coolwarm", interpolation='none')
plt.title("Correlation Matrix (imshow)")
plt.colorbar()
plt.xticks(range(len(df.select_dtypes(include='number').columns)), df.select_dtypes(include='number').columns, rotation=90)
plt.yticks(range(len(df.select_dtypes(include='number').columns)), df.select_dtypes(include='number').columns)
plt.tight_layout()
plt.show()

# Subplot Example
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].hist(df["CGPA"], color='blue')
axs[0, 0].set_title("Histogram of CGPA")

axs[0, 1].scatter(df["CGPA"], df["Salary_LPA"], color='green')
axs[0, 1].set_title("CGPA vs Salary")

axs[1, 0].bar(df["Gender"].value_counts().index, df["Gender"].value_counts().values)
axs[1, 0].set_title("Gender Count")

axs[1, 1].pie(placed_counts, labels=placed_counts.index, autopct="%1.1f%%")
axs[1, 1].set_title("Placement Ratio")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("student_dataset.csv")  

df_clean = df.dropna(subset=["Study_Hours_Per_Week", "CGPA"])

corr_coeff, p_value = pearsonr(df_clean["Study_Hours_Per_Week"], df_clean["CGPA"])

print(f"Pearson correlation coefficient between Study Hours and CGPA: {corr_coeff:.3f}")
print(f"P-value: {p_value:.3e}")

# Interpretation
if corr_coeff > 0.7:
    relation = "strong positive correlation"
elif corr_coeff > 0.3:
    relation = "moderate positive correlation"
elif corr_coeff > 0:
    relation = "weak positive correlation"
elif corr_coeff < -0.7:
    relation = "strong negative correlation"
elif corr_coeff < -0.3:
    relation = "moderate negative correlation"
elif corr_coeff < 0:
    relation = "weak negative correlation"
else:
    relation = "no correlation"

print(f"Interpretation: There is a {relation} between Study Hours and CGPA.")

# Plot scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x="Study_Hours_Per_Week", y="CGPA", data=df_clean, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Study Hours Per Week vs CGPA")
plt.xlabel("Study Hours Per Week")
plt.ylabel("CGPA")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create DataFrame from the given table
data = {
    "ID": [1, 2, None, 4, 5, 6, 7, 8, None, 10],
    "Age": [25, 37, None, 29, 42, 55, 23, 29, None, 45],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "Income": [50000, 62000, 45000, 54000, None, 100000, 52000, 70000, 200000, 85000],
    "Occupation": ["Engineer", "Data Scientist", None, "Product Manager", "Sales",
                   "Executive", None, "Product Manager", None, "Executive"],
    "Years_Employed": [2, 5, 3, None, 8, 30, 1, 6, 15, None],
    "Satisfaction_Level": [0.8, None, 0.6, 0.7, 0.5, 0.9, 0.4, 0.7, 0.9, None],
    "Purchase_History": ["High", "Medium", "Low", "High", "Low", "High", "Medium", "High", "Low", "High"]
}

df = pd.DataFrame(data)

# -----------------------------
# 1. Check missing values
print("Missing Values:\n", df.isnull().sum())

# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())   # Fill with mean
df['Income'] = df['Income'].fillna(df['Income'].median())  # Fill with median
df['Occupation'] = df['Occupation'].fillna("Unknown")      # Fill with "Unknown"
df['Years_Employed'] = df['Years_Employed'].fillna(df['Years_Employed'].mean())
df['Satisfaction_Level'] = df['Satisfaction_Level'].fillna(df['Satisfaction_Level'].mean())

# -----------------------------
# 2. Convert Satisfaction_Level  Binary (High if >0.7)
df['Satisfaction_Binary'] = df['Satisfaction_Level'].apply(lambda x: "High" if x > 0.7 else "Low")

# -----------------------------
# 3. Map Purchase_History  Numerical
purchase_map = {"Low": 0, "Medium": 1, "High": 2}
df['Purchase_History_Num'] = df['Purchase_History'].map(purchase_map)

# -----------------------------
# 4. Visualization

# Distribution of Satisfaction Level
plt.figure(figsize=(6,4))
sns.histplot(df['Satisfaction_Level'], bins=5, kde=True, color="blue")
plt.title("Distribution of Satisfaction Level")
plt.show()

# Barplot: Purchase history vs Income
plt.figure(figsize=(6,4))
sns.barplot(x="Purchase_History", y="Income", data=df, estimator="mean", errorbar=None)
plt.title("Average Income by Purchase History")
plt.show()

# Scatterplot: Age vs Income (colored by Satisfaction_Binary)
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Income", hue="Satisfaction_Binary", data=df, s=100)
plt.title("Age vs Income colored by Satisfaction")
plt.show()

# Heatmap: Correlation of numerical features
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
